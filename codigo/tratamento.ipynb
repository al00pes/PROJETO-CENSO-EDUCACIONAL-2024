# 1-  Objetivo

O objetivo dessa análise é buscar instituições privadas que podem tem um potencial 
alto para oferecemos os nosso produtos e dessa forma agregar valor a instutição.

Os dados que estou trabalhando são informações reais retiradas do site do INEP.
O dataset contém inumeras colunas, porém selecionei apenas 30 colunas que etão alinhadas com o que
desejo extrair desses dados.

As mais significativas são as que contém informações que seja possuim localizada 
a instuição , por meio de endereço fisico ou digital. Além disse possui outras informações relevantes
como a existência de biliotecas, equipamento de multimidia, computadores para os alunos , situação atual da instituições,
dentre outras.
